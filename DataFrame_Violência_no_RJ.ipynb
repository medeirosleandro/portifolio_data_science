##**É assustador o número de atentados contra a vida das mulheres, por ano, conforme vemos nos resultos abaixo**

##Total de FEMINICÍDIOS entre 2016 e 2021

```
soma_16 = df_feminicidio.loc[df_feminicidio['ano']== 2016].feminicidio.sum()
soma_17 = df_feminicidio.loc[df_feminicidio['ano']== 2017].feminicidio.sum()
soma_18 = df_feminicidio.loc[df_feminicidio['ano']== 2018].feminicidio.sum()
soma_19 = df_feminicidio.loc[df_feminicidio['ano']== 2019].feminicidio.sum()
soma_20 = df_feminicidio.loc[df_feminicidio['ano']== 2020].feminicidio.sum()
soma_21 = df_feminicidio.loc[df_feminicidio['ano']== 2021].feminicidio.sum()
total_fem  = df_feminicidio.feminicidio.sum()

print('FEMINICÍDIOS')
print('2016 =',soma_16)
print('2017 =',soma_17)
print('2018 =',soma_18)
print('2019 =',soma_19)
print('2020 =',soma_20)
print('2021 =',soma_21)
print('----------------')
print('Total: ',total_fem)
```

##TENTATIVAS  DE FEMINICÍDIO entre 2016 e 2021

```
soma_16T = df_feminicidio.loc[df_feminicidio['ano']== 2016].feminicidio_tentativa.sum()
soma_17T = df_feminicidio.loc[df_feminicidio['ano']== 2017].feminicidio_tentativa.sum()
soma_18T = df_feminicidio.loc[df_feminicidio['ano']== 2018].feminicidio_tentativa.sum()
soma_19T = df_feminicidio.loc[df_feminicidio['ano']== 2019].feminicidio_tentativa.sum()
soma_20T = df_feminicidio.loc[df_feminicidio['ano']== 2020].feminicidio_tentativa.sum()
soma_21T = df_feminicidio.loc[df_feminicidio['ano']== 2021].feminicidio_tentativa.sum()
total_ten_fem = df_feminicidio.feminicidio_tentativa.sum()

print('TENTATIVAS  DE FEMINICÍDIO')
print('2016 =',soma_16T)
print('2017 =',soma_17T)
print('2018 =',soma_18T)
print('2019 =',soma_19T)
print('2020 =',soma_20T)
print('2021 =',soma_21T)
print('----------------')
print('Total',total_ten_fem)
```

##Porcentagem de valores vazios nas colunas.

```
#Porcentagem valores vazios por coluna
vazios = df_rio.isnull().sum() / df_rj.shape[0]
print(round(vazios,2))
```

##Porcentagem de veiculos recuperados.
Com base na média de veiculos roubados e furtados, foi pssivel calcular a porcentagem de veículos recuperados no periodo de 1991 e 2019. 

```
#Media de Roubo de Veículos entre os anos de 1991 e 2019
media_roubo = df_rio.roubo_veiculo.mean()
print('Média anual de Roubo de veículos: {:.2f}'.format(round(media_roubo,2)))

#Média de Furto de Veículos entre os anos de 1991 e 2019
media_furto = df_rio.furto_veiculos.mean()
print('Média anual de Furtos de veículos: {:.2f}'.format(round(media_furto,2)))

total = media_furto + media_roubo
```

```
#Média de Recuperação de Veículos entre os anos de 1991 e 2019
print('Média de veículos recuperados:{:.2f}'.format(df_rio.recuperacao_veiculos.mean()))
```

```
#Porcentagem de veículos recuperados 
recuperado = df_rio.recuperacao_veiculos.mean()/ total*100
print('Porcentagem de veiculos recuperados :{:.2f}%'.format(recuperado))
```
