###Importando a Biblioteca Pandas

```
import pandas as pd
```

###Atribuindo o caminho do arquivo para uma constante

```
DATA_PATH = "https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/ocorrencias_aviacao.csv"
```

###Criando o DataFrame e usando como indice a coluna 'codigo_ocorrencia'


```
df = pd.read_csv(DATA_PATH,index_col='codigo_ocorrencia')
```

###Exibindo as 5 primeiras linhas do DataFrame


```
df.head()
```

###Informar a quantidade de Linhas e Colunas do DataFrame (Linhas, colunas)


```
df.shape
```

###Para exibir o resumo de todas as colunas do DataFrame, Quantidade de entradas,valores não-nulos, tipo de cada variável e quantidade de memória usada para carregar o DataFrame, podemos usar o método **info()**

```
df.info()
```

###Exibir apenas o Tipo das variáveis contidas no DataFrame


```
df.dtypes
```

###Exibe as ocorrências únicas de uma coluna (No exemplo foi usada a coluna ocorrencia_uf)


```
df.ocorrencia_uf.unique()
```

###Contagem de ocorrências por item

```
df.ocorrencia_uf.value_counts()
```

###Porcentagem de Ocorrências (Número de Ocorrências divido pelo numero de linhas)


```
df.ocorrencia_uf.value_counts() / df.shape[0]
```

###Para exibição de informações apenas de colunas específicas no DataFrame, usamos o método **loc**. No exemplo abaixo na primeira posição do Array informamos o indice e na segunda posição um Array com as colunas desejadas.

```
df.loc[201305055424986,['ocorrencia_latitude','ocorrencia_longitude']]
```

###Nos exemplos abaixo são feitos filtros por determinada informação nas colunas.

```
df.loc[df['ocorrencia_tipo']=='POUSO SEM TREM'].head()
```

```
df.loc[df['ocorrencia_classificacao']=='ACIDENTE'].head()
```

```
df.loc[df['ocorrencia_pais']=='BRASIL'].head()
```

```
df.loc[df['ocorrencia_tipo_categoria'] !='OUTROS'].head()
```

```
df.loc[df['ocorrencia_classificacao']=='INCIDENTE'].head()
```

###Para exibição de colunas informando o range de colunas específico, usamos o **.iloc** 
`loc[<numero_da_linha>, <numero_da_coluna>]`

```
df.iloc[:3,:3]
```

```
df.iloc[:1,:1]
```
