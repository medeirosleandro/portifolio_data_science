## <center>O que é o novo coronavírus? </center>
É um vírus que causa uma doença respiratória —a covid-19— pelo agente coronavírus, identificado em dezembro de 2019 na China. Os coronavírus são uma grande família viral, conhecidos desde meados de 1960, que causam infecções respiratórias em seres humanos e em animais. Geralmente, infecções por coronavírus causam doenças respiratórias leves a moderadas, semelhantes a um resfriado comum. Alguns coronavírus podem causar doenças graves com impacto importante em termos de saúde pública, como a Sars (Síndrome Respiratória Aguda Grave), identificada em 2002 e a Mers (Síndrome Respiratória do Oriente Médio), identificada em 2012.

##<center>Por que a doença foi batizada de covid-19?</center>

"Co" significa corona, "vi" vem de vírus, e "d" representa "doença". O número 19 indica o ano de sua aparição, 2019. Esse nome substitui o de 2019-nCoV, decidido provisoriamente após o surgimento da doença respiratória. O novo nome foi escolhido por ser fácil de pronunciar e não ter referência estigmatizante a um país ou a uma população em particular

```python
#PAÍSES COM MAIS MORTES

df_3= df.loc[df.date == '2021-05-13'].sort_values(by="total_deaths", ascending = False)
df_3 = df_3.drop(labels=[900,74409,27629,59261,4613,27152,87787], axis=0)
```

```python
deaths_top5 = df_3.loc[df.date == '2021-05-13', ['location','total_deaths']].sort_values(by='total_deaths',ascending=False)[:6]

sns.set()

fig, ax = plt.subplots(figsize=(10,5))

ax = sns.barplot(x='location', y='total_deaths', data=deaths_top5, palette="Blues_d")
ax.set_ylabel("Total de Mortes")
ax.set_xlabel("Localização")

plt.tight_layout()
```

```python
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
x = list(range(443))
z = df_brazil['total_cases']
z2 = df_brazil['total_deaths']

fig, ([ax1,ax2,ax3]) = plt.subplots(nrows=3, ncols=1, figsize=(10, 10))


ax1.set_title('Mortes no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Mortes')
ax1.plot(x, z2)

ax2.set_title('Mortes no Brasil (Escala logarítmica')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Mortes')
ax2.set_yscale('log')
ax2.plot(x, z2)


ax3.set_title('Casos no Brasil (Escala logarítmica)')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Casos')
ax3.set_yscale('log')
ax3.plot(x, z)

plt.tight_layout()

```
